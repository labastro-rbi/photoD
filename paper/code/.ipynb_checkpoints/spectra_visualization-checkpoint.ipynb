```python
import matplotlib.pyplot as plt
import numpy as np
import pysynphot as S
```

```python
teff_range = [3500., 50000.]
mh_range = [-3.,2.]
logg_range = [0, 6]
steps = (11, 11, 13)

pars = teff_range, mh_range, logg_range
```

```python
def plotter(g,m,t):
    plt.figure()
    sp = S.Icat('k93models', t, m, g)
    plt.semilogy(sp.wave, sp.flux, label=map(str, (t,m,g)))
    plt.xlim(1000, 15000)
    plt.xlabel(sp.waveunits)
    plt.ylabel(sp.fluxunits)
    plt.legend(loc='center right')
```

```python
for g in np.linspace(logg_range[0], logg_range[1], 7):
    for m in np.linspace(mh_range[0], mh_range[1], 11):
        for t in np.linspace(teff_range[0], teff_range[1], 11):
            print(g,m,t)
            try:
                plotter(g,m,t)
            except:
                pass
            input()
    
```
